### How many leading (n=1) roles were available to actors, and how many to actresses, in the 1950s?

```python
# Assuming the DataFrame has 'n' and 'year' columns
if 'n' in cast.columns and 'year' in cast.columns:
    # Filter rows for the 1950s
    roles_1950s = cast[(cast['year'] // 10 == 195)]

    # Count the number of leading roles (n=1) for actors and actresses
    leading_roles = roles_1950s[roles_1950s['n'] == 1]
    actors_count = leading_roles[leading_roles['type'] == 'actor'].shape[0]
    actresses_count = leading_roles[leading_roles['type'] == 'actress'].shape[0]

    # Display the results
    print(f"Number of leading (n=1) roles for actors in the 1950s: {actors_count}")
    print(f"Number of leading (n=1) roles for actresses in the 1950s: {actresses_count}")
else:
    print("The DataFrame does not contain 'n' or 'year' columns.")
```
